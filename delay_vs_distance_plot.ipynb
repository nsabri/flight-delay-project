```python
# necessary libraries
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# import data from csv file
df = pd.read_csv("flight_data_with_dist_and_dur.csv")
df.head()
```

```python
sns.set_theme(style="ticks", font_scale=1.25)
g = sns.relplot(
    data=df,
    x="flight_distance", y="target",
    s=30,
)
g.set_axis_labels("Flight distance [km]", "Delay [min]", labelpad=10)
g.figure.set_size_inches(10, 6)
g.ax.margins(.05)
g.despine(offset=10)
```

```python
sns.set_theme(style="ticks", font_scale=1.25)
g = sns.relplot(
    data=df,
    x="flight_duration", y="target",
    s=30,
)
g.set_axis_labels("Flight duration [min]", "Delay [min]", labelpad=10)
g.figure.set_size_inches(10, 6)
g.ax.margins(.05)
g.despine(offset=10)
```
