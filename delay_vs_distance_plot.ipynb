```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the training dataset
df = pd.read_csv("flight_data_with_dist_and_dur.csv")
df.head()
```

```python
# Set figure size and style
plt.figure(figsize=(12, 6))
sns.set_theme(style="ticks", font_scale=1.25)

g = sns.relplot(data=df, x="flight_duration", y="target", s=35, kind="scatter")
g.ax.margins(.05)
g.despine(offset=5)
g.figure.set_size_inches(12, 6)

# Set titles and labels
plt.title('Hallihallo Title', fontsize=24)
plt.xlabel('Flight duration [min]', fontsize=16, fontweight='bold')
plt.ylabel('Delay [min]', fontsize=16, fontweight='bold')

plt.show()
```
